```python
# =============================================
# ANALIZADOR DE PYMEs - v0.0
# Analiza la salud financiera de pequeñas
# y medianas empresas con Python básico
# =============================================


# --- BLOQUE 1: LISTA DE EMPRESAS ---
# Cada empresa es un diccionario con sus datos financieros
# La lista agrupa todas las empresas juntas para analizarlas de una vez

empresas = [
    {'nombre': 'Ferretería García', 'facturacion': 320000, 'coste_ventas': 210000, 'empleados': 5, 'deuda': 45000, 'activo_total': 180000},
    {'nombre': 'Bar El Rincón',     'facturacion': 180000, 'coste_ventas': 95000,  'empleados': 4, 'deuda': 12000, 'activo_total': 85000},
    {'nombre': 'Peluquería Ana',    'facturacion': 95000,  'coste_ventas': 38000,  'empleados': 2, 'deuda': 8000,  'activo_total': 40000},
]


# --- BLOQUE 2: ANÁLISIS CON LOOP ---
# El loop 'for' repite el análisis para cada empresa de la lista
# 'empresa' es el nombre temporal que le damos a cada elemento mientras lo procesamos

for empresa in empresas:

    # CÁLCULO DE RATIOS
    # Diferencia entre lo que ingresa y lo que le cuesta producir o comprar
    margen_bruto = empresa['facturacion'] - empresa['coste_ventas']

    # Qué proporción de lo que tiene la empresa es deuda (0.25 = 25% de deuda)
    ratio_endeudamiento = empresa['deuda'] / empresa['activo_total']

    # Cuánto genera cada empleado — mide la eficiencia del negocio
    fact_por_empleado = empresa['facturacion'] / empresa['empleados']

    # SCORING DE RIESGO
    # Clasifica la empresa según su nivel de deuda respecto a sus activos
    if ratio_endeudamiento > 0.6:
        riesgo = 'ALTO'    # más del 60% de lo que tiene es deuda
    elif ratio_endeudamiento > 0.3:
        riesgo = 'MEDIO'   # entre el 30% y el 60%
    else:
        riesgo = 'BAJO'    # menos del 30% — empresa saneada

    # MOSTRAR RESULTADOS
    # Imprime los resultados de esta empresa antes de pasar a la siguiente
    print('---', empresa['nombre'])
    print('Margen bruto:', margen_bruto, '€')
    print('Ratio endeudamiento:', round(ratio_endeudamiento, 2))
    print('Facturación por empleado:', round(fact_por_empleado, 0), '€')
    print('Empresa:', empresa['nombre'])
    print('Ratio endeudamiento:', round(ratio_endeudamiento, 2))
    print('Nivel de riesgo:', riesgo)
```
